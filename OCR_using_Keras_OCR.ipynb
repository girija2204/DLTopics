```
import matplotlib.pyplot as plt
import keras_ocr
```

```
pipeline = keras_ocr.pipeline.Pipeline()
```

```
images = [keras_ocr.tools.read(image) for image in ["https://i.imgur.com/9C5Q9Rt.jpg","https://i.pinimg.com/736x/0f/50/54/0f50543167ea6284153c6282918715c4.jpg","https://upload.wikimedia.org/wikipedia/commons/thumb/0/0b/ReceiptSwiss.jpg/170px-ReceiptSwiss.jpg"]]
```

```
prediction = pipeline.recognize(images)
```

```
prediction[0]
```

```
fig,axs = plt.subplots(nrows=len(images), figsize=(20,20))
for ax, image, predictions in zip(axs, images, prediction):
  keras_ocr.tools.drawAnnotations(image=image,predictions=predictions,ax=ax)
```
